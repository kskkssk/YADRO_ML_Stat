# Задача 1

Пусть $X_1, X_2, \ldots, X_n$ — выборка из экспоненциального распределения с параметром $\lambda$. Найти оценку максимального правдоподобия параметра $\lambda$, сравнить ее с байесовской оценкой (MAP и математическое ожидание апостреорного распределения), подобрав сопряженное распределение. Сравнить полученные байесовские оценки с оценкой MLE. Найти предсказательное распределение

## MLE

$$
Likehood(D, \lambda) = ∏^n_{i=1}P(x_i|\lambda) = ∏^n_{i=1}\lambda e^{-\lambda x_i} = \lambda^ne^{-\lambda n \overline{x}}
$$

$$
LnLikehood(D, \lambda) = ln(\lambda^ne^{-\lambda n \overline{x}}) = nln(\lambda) -\lambda n \overline{x}
$$

$$
\frac{\partial{LnLikehood(D, \lambda)}}{\partial \lambda} = \frac{n}{\lambda} - n\overline{x}
$$

Тогда,
$$
\lambda_{MLE} = \frac{1}{\overline{x}} = \frac{n}{\sum^n_{i=1}x_i}
$$

## Апостериорное распределение

$$
P(\lambda|D) = \frac{P(D|\lambda)P(\lambda)}{P(D)}
$$
$$
P(D|\lambda) = \lambda^ne^{-\lambda n \overline{x}}
$$
$$
P(\lambda) = \frac{\lambda^{a-1}β^ae^{-\beta\lambda}}{Г(a)}
$$
$$
P(D) = \int^{\infty}_0P(D|\lambda)P(\lambda)d\lambda = \frac{β^a}{Г(a)} \int^{\infty}_0\lambda^{n + a - 1}e^{-\lambda(\beta + n\overline{x})}\lambda = \frac{β^a \cdot Г(n + a)}{Г(a) \cdot (β + n\overline{x})^{n + a}} \int^{\infty}_0Gamma(n+a, \beta+n\overline{x})d\lambda = \frac{β^a \cdot Г(n + a)}{Г(a) \cdot (β + n\overline{x})^{n + a}}
$$

Тогда

$$
P(\lambda|D) = \frac{\lambda^{n + a - 1}e^{-\lambda(\beta + n\overline{x})}(\beta + n\overline{x})^{n + a}}{Г(n + a)} = Gamma(n+a, \beta + n \overline{x})
$$

Найдем MAP как моду Гамма-распределения и математическое ожидание апостреорного распределения как математическое ожидание Гамма-распределения.

$$
\lambda_{MAP} = \frac{n + a - 1}{\sum^n_{i=1}x_i + \beta}
$$

$$
\lambda_{E} = \frac{n + a}{\sum^n_{i=1}x_i + \beta}
$$

## Предсказательное распределение

$$
P(X|D) = \int^{\infty}_0P(X|\lambda)P(\lambda|D)d\lambda
$$

$$
P(X|\lambda) = \lambda e^{-\lambda x}
$$

$$
P(\lambda|D) = Gamma(n+a, \beta + n \overline{x})
$$

Тогда

$$
P(X|D) = \int^{\infty}_0P(X|\lambda)P(\lambda|D)d\lambda =
\int^{\infty}_0\frac{\lambda e^{-\lambda x} \cdot \lambda^{n + a - 1}e^{-\lambda(\beta + n\overline{x})}(\beta + n\overline{x})^{n + a}}{Г(n + a)}d\lambda =
\int^{\infty}_0\frac{\lambda^{n + a}e^{-\lambda(\beta + \sum^{n+1}_{i=0}x_i)}(\beta + \sum^{n}_{i=0}x_i))^{n + a}}{Г(n + a)}d\lambda
= \frac{(\beta + \sum^{n}_{i=0}x_i))^{n + a}}{{Г(n + a)}}\int^{\infty}_0\lambda^{n + a}e^{-\lambda(\beta + \sum^{n+1}_{i=0}x_i)}d\lambda
= \frac{(\beta + \sum^{n}_{i=0}x_i))^{n + a}}{{Г(n + a)}} \cdot \frac{Г(n + a + 1)}{(\beta + \sum^{n+1}_{i=0}x_i)^{n + a + 1}}\int^{\infty}_0\lambda^{n + a}e^{-\lambda(\beta + \sum^{n+1}_{i=0}x_i)}d \cdot \frac{(\beta + \sum^{n+1}_{i=0}x_i)^{n + a + 1}}{Г(n + a + 1)}d\lambda
= \frac{(\beta + \sum^{n}_{i=0}x_i))^{n + a}}{{Г(n + a)}} \cdot \frac{Г(n + a + 1)}{(\beta + \sum^{n+1}_{i=0}x_i)^{n + a + 1}}\int^{\infty}_0 Gamma(n + a + 1,\beta + \sum^{n+1}_{i=0}x_i)d\lambda
=  \frac{(\beta + \sum^{n}_{i=0}x_i))^{n + a}}{{Г(n + a)}} \cdot \frac{Г(n + a + 1)}{(\beta + \sum^{n+1}_{i=0}x_i)^{n + a + 1}}
= \frac{(\beta + \sum^{n}_{i=0}x_i))^{n + a}}{(\beta + \sum^{n+1}_{i=0}x_i)^{n + a + 1}} \cdot (n + a)
$$

# Задача 2

**Мультиномиальное распределение**

Пусть проводится серия из $n$ испытаний и в результате каждого испытания происходит ровно одно событие из набора $A_1, A_2, \dots, A_m$, причем вероятности этих событий равны соответственно $\mathsf{p}_1, \mathsf{p}_2, \dots, \mathsf{p}_m$, причем
$$\sum_{i=1}^{m}\mathsf{p}_i = 1.$$

Тогда совместное распределение величин $X_1, X_2, \dots, X_m$, где $X_k$ — число наступлений события $A_k$ в серии из $n$ испытаний, задается вероятностями

$$
\mathsf{P}\left(X_1 = n_1, \dots, X_m = n_m, \right) = \frac{n!}{n_1!\dots n_m!}\mathsf{p}_1^{n_1}\dots \mathsf{p}_m^{n_m},
$$

где $n_1, n_2, \dots, n_m$ — произвольный набор целых неотрицательных чисел, таких что

$$\sum_{i=1}^m n_i = n.$$

Произведите байесовский вывод для мультиномиального распределения: найдите апостериорное распределение, используя в качестве сопоряженного распределения к правдоподобию [распределение Дирихле](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%94%D0%B8%D1%80%D0%B8%D1%85%D0%BB%D0%B5), найдите предсказательное распределение. Объясните результат.

$$
P(\lambda|D) = \frac{P(D|\lambda)P(\lambda)}{P(D)}
$$

$$
P(D|p) = \frac{n!}{X_1! \dots X_m!} p_1^{X_1} \dots p_m^{X_m}
$$

$$
P(p) = \frac{1}{B(a_1 \dots a_m)} \prod^m_{i=1}p_i^{a_i - 1} =
\frac{Г(a)p_1^{a_1-1} \dots p_m^{a_m-1} }{\prod^m_{i=1}Г(a_i)}
$$

$$
P(X_1, \dots, X_m) = \int P(p)P(D|p)dp = \int \frac{Г(a)\prod^m_{i=1}p_i^{a_i - 1}}{\prod^m_{i=1}Г(a_i)} \cdot \frac{n!}{X_1! \dots X_m!} p_1^{X_1} \dots p_m^{X_m} =
\int (\frac{n!}{X_1! \dots X_m!}\frac{Г(a)}{\prod^m_{i=1}Г(a_i)}\frac{\prod^m_{i=1}Г(X_i + a_i)}{Г(n + a)}) \cdot \frac{Г(n + a)p_1^{X_1 + a_1 - 1} \dots p_m^{X_m + a_m - 1}}{\prod^m_{i=1}Г(X_i + a_i)}dp =
\frac{n!}{X_1! \dots X_m!}\frac{Г(a)}{\prod^m_{i=1}Г(a_i)}\frac{\prod^m_{i=1}Г(X_i + a_i)}{Г(n + a)} =
\frac{n!}{X_1! \dots X_m!} \frac{(a-1)!}{(a_1-1)! \dots (a_m-1)!} \frac{(X_i +a_1 - 1)! \dots (X_m + a_m - 1)!}{(n + a - 1)!}
$$

Тогда

$$
P(p|D) = \frac{(n + a - 1)!}{(X_i +a_1 - 1)! \dots (X_m + a_m - 1)!} p_1^{X_1 + a_1 -1} \dots p_m^{X_m + a_m - 1} = \frac{Г(n + a)}{\prod^m_{i=1}Г(X_i +a_i)} p_1^{X_1 + a_1 -1} \dots p_m^{X_m + a_m - 1}
$$

Это распределение Дерихле с $α = (X_1 + a_1, \dots X_m + a_m)$.


Предскательное распределение:

$$
P(Y = y|D) = \int P(Y|p)P(p|D)dp = \int\frac{k!}{Y_1! \dots Y_m!} p_1^{Y_1} \dots p_m^{Y_m} \cdot \frac{(n + a - 1)!}{(X_1 +a_1 - 1)! \dots (X_m + a_m - 1)!} p_1^{X_1 + a_1 -1} \dots p_m^{X_m + a_m - 1}dp
= \frac{k!(n+a - 1)!}{Y_1! \dots Y_m!(X_1 +a_1 - 1)! \dots (X_m + a_m - 1)!}\int p_1^{Y_1 + X_1 +a_1 - 1} \dots p_1^{Y_m + X_m +a_m - 1}dp
=\frac{k!(n+a-1)!}{Y_1! \dots Y_m!(X_1 +a_1-1)! \dots (X_m + a_m-1)!} \cdot \frac{(Y_1 + X_1 +a_1-1)! \dots (Y_m + X_m +a_m-1)!}{(k + n + a-1)!}\int \frac{(k + n + a-1)!}{(Y_1 + X_1 +a_1-1)! \dots (Y_m + X_m +a_m-1)!} p_1^{Y_1 + X_1 +a_1 - 1} \dots p_1^{Y_m + X_m +a_m - 1}dp
= \frac{k!(n+a-1)!}{Y_1! \dots Y_m!(X_i +a_1-1)! \dots (X_m + a_m-1)!} \cdot \frac{(Y_1 + X_1 +a_1-1)! \dots (Y_m + X_m +a_m-1)!}{(k + n + a-1)!}
$$

$$
P(Y = y|D) = \frac{k!Г(n+a)}{\prod^m_{i=1}Y_i!Г(X_i + a_i)} \cdot \frac{\prod^m_{i=1}Г(Y_i + X_i + a_i)}{Г(k + n + a)}
$$
