#### **1. Определить (с обоснованием), зависимы или независимы следующие события:**

(a) Несовместные события;

(b) События, образующие σ-алгебру Σ в пространстве (Ω, Σ, P);

(c) События, имеющие одинаковую вероятность?


**(a) Вероятность несовместных событий равна 0.**

Как зависимы, так и независимы

P(A ∩ B) = P(A) * P(B) = 0, если P(A) = 0 | P(B) = 0.

События независимы, если вероятность одного из них = 0.

Зависимы, если P(A) > 0 & P(B) > 0

**(b) События, образующие σ-алгебру Σ в пространстве (Ω, Σ, P).**

Как зависимы, так и независимы.

Зависимые пары - событие A и его дополнение A'. P(A ∩ A') = 0 ≠ P(A) * P(A')

Независимые пары - P(A ∩ ∅) = 0 = P(A) * P(∅)

**(c) События, имеющие одинаковую вероятность?**

Как зависимы, так и независимы

Условие P(A) = P(B) не дает сведений о зависимости событий

#### **4. Пусть ξ ∼ U−π,π и η1 = cos ξ, η2 = sin ξ.**

Плотность распределения: f(x) = 1 / Pi - (-Pi) = 1 / 2Pi

**Мат. ожидание непрерывной случайной величины**

E[x] = $\int_{-\Pi}^{\Pi} x * f(x)\, dx$

E[ξ] = $\int_{-\Pi}^{\Pi} x * 1/2Pi\, dx$

E[ξ] = 1/2Pi * $\int_{-\Pi}^{\Pi} x\, dx$ = 1/2Pi * [x^2/2](Pi, -Pi) = 1/2Pi * (Pi^2/2 - (-Pi)^2/2) = 1/2Pi * 0 = 0

E[η1] = $\int_{-\Pi}^{\Pi} cos(x) * 1/2Pi\, dx$

E[η1] = 1/2Pi * $\int_{-\Pi}^{\Pi} cos(x)\, dx$ = 1/2Pi * [sin(x)](Pi, -Pi) = 1/2Pi * (sin(Pi) - sin(-Pi)) = 1/2Pi * 0 = 0

E[η2] = $\int_{-\Pi}^{\Pi} sin(x) * 1/2Pi\, dx$

E[η2] = 1/2Pi * $\int_{-\Pi}^{\Pi} sin(x)\, dx$ = 1/2Pi * [-cos(x)](Pi, -Pi) = 1/2Pi * (-cos(Pi) - (-cos(-Pi))) = 1/2Pi * (1 - 1) = 0

**cov(ξ, η1) = 0**

**cov(η1, η2) = 0**

#### **5. Найти плотность распределения суммы двух независимых случайных величин ξ и η, если ξ ∼ Exp2 и η ∼ U0,1.**

Пусть случайные величины ξ  η, абсолютно непрерывны с плотностями pξ(x), pη(y) и независимы. Тогда получаем формулу свертки:

**pξ + pη(z) = $\int_{-\inf}^{\inf} pξ(x) * pη(z - x)\, dx$**

**Распределения:**

1. Показательное $f_\xi(x)$ = λe^(-λx), x>=0

2. Равномерное $f_\eta(y)$ = 1, a <= y <= b, a = 0, b = 1

**Плотность распределения:**

1. Показательное $f_\xi(x) = 2 \cdot e^{-2x}$, x>=0

2. Равномерное $f_\eta(y) = 1$, 0 <= y <= 1

$f_\xi(x)$: x >= 0

$f_\eta(y)$: x <= z; x >= z - 1

Случай: 0 <= z <= 1:

x >= z - 1 (Автоматически выполняется при x >= 0)

pξ + pη(z) = $\int_{0}^{z} (2e^{-2x}) \cdot 1\, dx$ = -e^-2z + 1 = 1 - e^-2z

Случай: z > 1:

Выполняется x >= z - 1. Автоматически выполняется x >= 0

pξ + pη(z) = $\int_{z - 1}^{z} (2e^{-2x}) \cdot 1\, dx$ = -e^-2z + -e^-2(
z-1)

Плотность: $f(z) = \begin{cases} 1 - e^{-2z}, & 0 \leq z \leq 1 \\ -e^{-2z} + -e^{-2(z-1)}, & z > 1 \\ 0, & z < 0 \end{cases}$
